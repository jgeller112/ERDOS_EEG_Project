```python
from scipy import signal # after some research I think a periodogram will work better
```

```python
def meanfreq(x, fs, subject_name):
    """
    Compute the power spectral density and mean frequency of the given subject over all trials.
    ----------
    x : array_like
        Time series of measurement values
    fs : float
        Sampling frequency in Hz.
    subj_id : string
        the subject identifier.
    Returns:
    subj_id : the subjects name.
    mnfreq : the mean frequency.
    """
    
    subj_id = subject_name
    f, Pxx_den = signal.periodogram(x, fs)                                                    
    Pxx_den = np.reshape( Pxx_den, (1,-1) ) 
    width = np.tile(f[1]-f[0], (1, Pxx_den.shape[1]))
    f = np.reshape(f, (1, -1))
    P = Pxx_den * width
    pwr = np.sum(P)

    mnfreq = np.dot(P, f.T)/pwr
    
    return list([subj_id,mnfreq])
```
