```python
## Plot the subject features
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## reading dataframe
df = pd.read_csv('./data/output_250_550ms/SubjectFeatureMeanAcrossIndividualChannels_S1obj.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(18,9))
for i in range(0,np.shape(df)[0]-1,2):
    #print(i)
    for j in range(i,np.shape(df)[0]-1,2):
        if (i != j):
            if (df.iloc[i,1] == 'a') & (df.iloc[j,1] == 'a'):
                #plt.scatter(df.iloc[i,3:-1], df.iloc[j,3:-1], color="black", marker="o", facecolor="black")
                plt.scatter(df.iloc[i,20:21], df.iloc[j,20:21], color="black", marker="o", facecolor="black")
            elif (df.iloc[i,1] == 'c') & (df.iloc[j,1] == 'c'):
                #plt.scatter(df.iloc[i,3:-1], df.iloc[j,3:-1], color="blue", marker="s", facecolor="none")
                plt.scatter(df.iloc[i,20:21], df.iloc[j,20:21], color="blue", marker="s", facecolor="none")
```

```python
dfA = df.loc[df['group'] == 'a']
dfC = df.loc[df['group'] == 'c']
```

```python
dfA['group'].unique()
```

```python
sensorValueA = np.zeros([np.shape(dfA)[0] , 64 ])
sensorValueC = np.zeros([np.shape(dfC)[0] , 64 ])
for i in range(0,np.shape(dfA)[0],1):
    sensorValueA[i,:] = dfA.iloc[i,3:]

for i in range(0,np.shape(dfC)[0],1):
    sensorValueC[i,:] = dfC.iloc[i,3:]
```

```python
figureName='./Figures/HistogramSensorVoltage.pdf'
plt.figure(figsize=(10,8))
#x=plt.hist(sensorValueA[:], bins=range(-20,20,1), density=False, facecolor="red" )
#y=plt.hist(sensorValueC[:], bins=range(-20,20,1), density=False, facecolor="blue" )
plt.hist(sensorValueA.ravel(), bins=25, density=False, facecolor="red" , histtype="step", linewidth=3, label="a" )
plt.hist(sensorValueC.ravel(), bins=25, density=False, facecolor="blue" , histtype="step", linewidth=3, label="c")
plt.legend(loc="best", fontsize=30)
#plt.gca().set_xticks([])
plt.gca().set_yticks([])
plt.gca().tick_params(labelsize=30)
plt.xlabel('Voltage (mV)', fontsize=30)
plt.savefig(figureName, format='pdf', bbox_inches='tight')
```
